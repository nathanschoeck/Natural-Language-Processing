1. Import Libraries

```python
from transformers import MarianMTModel, MarianTokenizer
```

2. Translate to Egnlish Function

```python
def translate_to_english(texts):
    """
    Translate a list of Spanish texts into English using the MarianMT model.

    Parameters:
    - texts (list of str): Sentences in Spanish to be translated.

    Returns:
    - List of translated sentences in English.
    """
    # Load the MarianMT model for Spanish to English translation
    model_name = "Helsinki-NLP/opus-mt-es-en"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    # Tokenize and translate the input texts
    inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True)
    translated = model.generate(**inputs)

    # Decode the translated output
    translated_texts = [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
    return translated_texts
```

3. Main Function with Spanish Sentences, Translation call, and Results Display

```python
if __name__ == "__main__":
    # Example sentences in Spanish
    spanish_texts = [
        "Hola, ¿cómo estás?",
        "Disfruto aprendiendo nuevos idiomas.",
        "Este es un ejemplo de traducción automática.",
        "¿Puedes ayudarme con esta tarea?",
        "Me encanta programar y explorar la tecnología."
    ]

    # Translate to English
    translated_texts = translate_to_english(spanish_texts)

    # Display the results
    for i, (original, translated) in enumerate(zip(spanish_texts, translated_texts), 1):
        print(f"{i}. Spanish: {original}")
        print(f"   English: {translated}")
        print("-" * 50)
```
